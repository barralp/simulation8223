# Differential equation

In the double pendulum problem we got from the Lagrangian:
\begin{equation}
\begin{array}{rll}
3\frac{d^{2}\theta_{1}}{dt^{2}} + 2\frac{d^{2}\theta_{2}}{dt^{2}}\cos(\theta_{1}-\theta_{2}) + 2\left(\frac{d\theta_{2}}{dt}\right)^{2}\sin(\theta_{1}-\theta_{2}) +\frac{g}{l}\sin\theta_{1} &=& 0 \\
4\frac{d^{2}\theta_{2}}{dt^{2}} + 3\frac{d^{2}\theta_{1}}{dt^{2}}\cos(\theta_{1}-\theta_{2}) - 3\left(\frac{d\theta_{1}}{dt}\right)^{2}\sin(\theta_{1}-\theta_{2}) + \frac{g}{l}\sin\theta_{2} &=& 0 \\
\end{array}
\end{equation}

Which gets transformed into this first order system:
\begin{equation}
\begin{array}{rll}
\frac{d\theta_{1}}{dt} &=& \Omega_{1} \\
\frac{d\theta_{2}}{dt} &=& \Omega_{2}  \\
3\frac{d\Omega_{1}}{dt} + 2\frac{d\Omega_{2}}{dt}\cos(\theta_{1}-\theta_{2}) + 2\Omega_{2}^{2}\sin(\theta_{1}-\theta_{2}) +\frac{g}{l}\sin\theta_{1} &=& 0 \\
4\frac{d\Omega_{2}}{dt} + 3\frac{d\Omega_{1}}{dt}\cos(\theta_{1}-\theta_{2}) - 3\Omega_{1}^{2}\sin(\theta_{1}-\theta_{2}) + \frac{g}{l}\sin\theta_{2} &=& 0 \\
\end{array}
\end{equation}

We want to use dimensionless quantities. $\theta_{i}$ are already dimensionless. We now only have the replace the time $t = T\tau$ with $T = 2\pi\sqrt{\frac{l}{g}}$. And we write $\tilde{\Omega}_{i} = T\Omega_{i}$. The equations become:

It leads to the following system:
\begin{equation}
\begin{array}{rllll}
\frac{d\theta_{1}}{d\tau} &=& \tilde{\Omega}_{1}\\
\frac{d\theta_{2}}{d\tau} &=& \tilde{\Omega}_{2}\\
3\frac{d\tilde{\Omega_{1}}}{d\tau} + 2\frac{\tilde{d\Omega_{2}}}{d\tau}\cos(\theta_{1}-\theta_{2}) + 2\tilde{\Omega}_{2}^{2}\sin(\theta_{1}-\theta_{2}) +4\pi^{2}\sin\theta_{1} &=& 0 \\
4\frac{d\tilde{\Omega}_{2}}{d\tau} + 3\frac{d\tilde{\Omega}_{1}}{d\tau}\cos(\theta_{1}-\theta_{2}) - 3\tilde{\Omega}_{1}^{2}\sin(\theta_{1}-\theta_{2}) + 4\pi^{2}\sin\theta_{2} &=& 0 \\
\end{array}
\end{equation}

However we need to keep the derivatives only on one side of the equation to integrate. We can do $2(3)-(4)\cos(\theta_{1}-\theta_{2})$ and $(3)\cos(\theta_{1}-\theta_{2})- (4)$ to obtain:

It leads to the following system:
\begin{equation}
\begin{array}{rllll}
\frac{d\theta_{1}}{d\tau} &=& \tilde{\Omega}_{1}\\
\frac{d\theta_{2}}{d\tau} &=& \tilde{\Omega}_{2}\\
\frac{d\tilde{\Omega_{1}}}{d\tau}(6-3\cos^{2}(\theta_{1}-\theta_{2})) + \sin(\theta_{1}-\theta_{2})(4\tilde{\Omega}_{2}^{2}+3\tilde{\Omega}_{1}^{2}\cos(\theta_{1}-\theta_{2})) +4\pi^{2}(2\sin\theta_{1}-\cos(\theta_{1}-\theta_{2})\sin\theta_{2}) &=& 0 \\
\frac{d\tilde{\Omega}_{2}}{d\tau}(2\cos^{2}(\theta_{1}-\theta_{2})-4) + \sin(\theta_{1}-\theta_{2})(2\tilde{\Omega}_{2}^{2}\cos(\theta_{1}-\theta_{2})+3\tilde{\Omega}_{1}^{2}) + 4\pi^{2}(\cos(\theta_{1}-\theta_{2})\sin\theta_{1}-\sin\theta_{2})&=& 0 \\
\end{array}
\end{equation}

Putting all the derivative on the same side we need to integrate:
\begin{equation}
\begin{array}{rllll}
\frac{d\theta_{1}}{d\tau} &=& \tilde{\Omega}_{1}\\
\frac{d\theta_{2}}{d\tau} &=& \tilde{\Omega}_{2}\\
\frac{d\tilde{\Omega_{1}}}{d\tau} &=& -\frac{\sin(\theta_{1}-\theta_{2})(4\tilde{\Omega}_{2}^{2}+3\tilde{\Omega}_{1}^{2}\cos(\theta_{1}-\theta_{2})) +4\pi^{2}(2\sin\theta_{1}-\cos(\theta_{1}-\theta_{2})\sin\theta_{2})}{(6-3\cos^{2}(\theta_{1}-\theta_{2}))} &=& f_{1}(\theta_{1}, \theta_{2},\tilde{\Omega}_{1}, \tilde{\Omega}_{2}) \\
\frac{d\tilde{\Omega}_{2}}{d\tau} &=& - \frac{\sin(\theta_{1}-\theta_{2})(2\tilde{\Omega}_{2}^{2}\cos(\theta_{1}-\theta_{2})+3\tilde{\Omega}_{1}^{2}) + 4\pi^{2}(\cos(\theta_{1}-\theta_{2})\sin\theta_{1}-\sin\theta_{2})}{(2\cos^{2}(\theta_{1}-\theta_{2})-4)}) &=& f_{2}(\theta_{1}, \theta_{2},\tilde{\Omega}_{1}, \tilde{\Omega}_{2}) \\
\end{array}
\end{equation}

```python
# coding: utf-8

import numpy as np
import matplotlib.pyplot as plt

def f_1(theta1, theta2, Omega1, Omega2):    # derivate of Omega_1
    return -(np.sin(theta1-theta2)*(4*Omega2**2+3*Omega1**2*np.cos(theta1-theta2))+4*np.pi**2*(2*np.sin(theta1)-np.cos(theta1-theta2)*np.sin(theta2)))/(6-3*(np.cos(theta1-theta2))**2)

def f_2(theta1, theta2, Omega1, Omega2):    # derivate of Omega_2
    return -(np.sin(theta1-theta2)*(2*Omega2**2*np.cos(theta1-theta2)+3*Omega1**2)+4*np.pi**2*(np.cos(theta1-theta2)*np.sin(theta1)-np.sin(theta2)))/(2*(np.cos(theta1-theta2))**2-4)

# Code to plot the vector field
thetaMax = 2*np.pi
thetaX = np.linspace(-thetaMax, thetaMax, 20)
thetaY = np.linspace(-thetaMax, thetaMax, 20)

u, v = np.meshgrid(thetaX, thetaY)

U = f_1(u, v, 0, 0)
V = f_2(u, v, 0, 0)


fig1, ax1 = plt.subplots(figsize = (10,10))
ax1.set_title('Vector field at zero velocity')
ax1.set(xlabel=r'$\theta_{1}/\pi$', ylabel=r'$\theta_{2}/\pi$')
Q = ax1.quiver(u/(np.pi), v/(np.pi), U, V, units='width')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
def advance(theta1, theta2, Omega1, Omega2, t, dt):   # theta1, theta2, Omega1, Omega2 are arrays, t total time of the simulation, dt time step
    t_index = int(round(t/dt))
    theta1[t_index+1] = theta1[t_index] + Omega1[t_index]*dt
    theta2[t_index+1] = theta2[t_index] + Omega2[t_index]*dt
    Omega1[t_index+1] = Omega1[t_index] + f_1(theta1[t_index],theta2[t_index], Omega1[t_index], Omega2[t_index])*dt
    Omega2[t_index+1] = Omega2[t_index] + f_2(theta1[t_index],theta2[t_index], Omega1[t_index], Omega2[t_index])*dt



def simulate(theta1_0, theta2_0, Omega1_0 = 0, Omega2_0 = 0, Tmax = 100, dt = 0.001):
    #dt = 0.05
    size_arr = int(Tmax/dt)+1
    T = np.linspace(0,Tmax,size_arr)
    theta1 = np.zeros(size_arr)
    theta2 = np.zeros(size_arr)
    Omega1 = np.zeros(size_arr)
    Omega2 = np.zeros(size_arr)
    theta1[0] = theta1_0
    theta2[0] = theta2_0
    Omega1[0] = Omega1_0
    Omega2[0] = Omega2_0

    for t in T[:-1]:
        advance(theta1, theta2, Omega1, Omega2, t, dt)
    
    return theta1, theta2, Omega1, Omega2

theta1, theta2, Omega1, Omega2 = simulate(theta1_0 = np.pi/10, theta2_0 = np.pi/3, Tmax = 3)
thetaMax = 2*np.pi

fig1, ax1 = plt.subplots(figsize = (10,10))
ax1.set_title(r'Simulation for , $\theta_{1,0} = \pi/10$, $\theta_{2,0} = \pi/3$')
ax1.set(xlabel=r'$\theta_{1}/\pi$', ylabel=r'$\theta_{2}/\pi$')
Q = ax1.quiver(u/np.pi, v/np.pi, U, V, units='width')
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(theta1/np.pi, theta2/np.pi, 'ro', ms = 0.1)
plt.xlim(-thetaMax/np.pi,thetaMax/np.pi)
plt.ylim(-thetaMax/np.pi,thetaMax/np.pi)
plt.show()
```

```python
def vectorField(thetaMax = 2*np.pi):
    thetaX = np.linspace(-thetaMax, thetaMax, 20)
    thetaY = np.linspace(-thetaMax, thetaMax, 20)
    u, v = np.meshgrid(thetaX, thetaY)
    U = f_1(u, v, 0, 0)
    V = f_2(u, v, 0, 0)
    return u, v, U, V
```

```python
def plot_pendulum(axis, theta1_0, theta2_0):
    r1 = np.linspace(0,3,50)
    r2 = np.linspace(0,4,50)
    x1 = 3*np.sin(theta1_0)
    y1 = -3*np.cos(theta1_0)
    x2 = x1 + 4*np.sin(theta2_0)
    y2 = y1 - 4*np.cos(theta2_0)
    bar1 = np.array([r1*np.sin(theta1_0), -r1*np.cos(theta1_0)])
    bar2 = np.array([x1 + r2*np.sin(theta2_0), y1-r2*np.cos(theta2_0)])
    axis.plot([x1, x2], [y1, y2], 'bo', ms = 3)
    axis.plot(bar1[0,:], bar1[1,:], '-', color = 'black')
    axis.plot(bar2[0,:], bar2[1,:], '-', color = 'red')
```

```python
import matplotlib.animation as animation
from IPython.display import HTML

def animate_double_pendulum(data, Tmax, title = '8233_animation.mp4', save = 'False'):
    # ajouter le nom des variables
    #setup figure
    theta1, theta2, Omega1, Omega2 = data
    fig = plt.figure(figsize = (10,10))
    ax1 = plt.subplot2grid((2,2), (0,0))
    ax2 = plt.subplot2grid((2,2), (0,1))
    ln, = plt.plot([],[])
    
    def init():
        thetaMax = 2*np.pi
        u, v, U, V = vectorField(thetaMax)
        
        ax1.set_title(r'$\theta_{1}, \theta_{2}$ coordinates')
        ax1.set(xlabel=r'$\theta_{1}/\pi$', ylabel=r'$\theta_{2}/\pi$')
        Q = ax1.quiver(u/(np.pi), v/(np.pi), U, V, units='width')
        plt.gca().set_aspect('equal', adjustable='box')
        ax1.set_xlim(-thetaMax/np.pi,thetaMax/np.pi)
        ax1.set_ylim(-thetaMax/np.pi,thetaMax/np.pi)
        ax1.plot(theta1/np.pi, theta2/np.pi, 'bo', ms = 0.01)
        #ax1.set_aspect('equal')
        plt.axis('scaled')
        ax2.set_xlim(-7.5,7.5)
        ax2.set_ylim(-7.5,7.5)
        #ax2.hist(np.zeros(nrep), bins= 20, range = Range[0])
        return ln,
    
    def animate(frame):
        title = '$T = ' + str(round(frame/(len(theta1))*Tmax, 2)) +'$'
        plt.title(title)
        #ax1.cla()
        ax2.cla()
        ax2.set_title(r'Pendulum animation')
        ax2.set_xlim(-7.5,7.5)
        ax2.set_ylim(-7.5,7.5)
        #ax2.set_aspect('equal')
        ax1.plot(theta1[frame]/np.pi,theta2[frame]/np.pi, 'ro')
        plot_pendulum(ax2, theta1[frame], theta2[frame])
        return ln,

    ani = animation.FuncAnimation(fig, animate, init_func = init, frames = range(Tmax), blit = True)
    if save:
        ani.save(title, writer=animation.FFMpegFileWriter(), dpi = 200)
    plt.close()
    return ani
```

```python
def dropFrame(simulationList, ratio):
    return simulationList[list(range(0, simulationList.shape[0], ratio))]

def simulateDropped(theta1_0, theta2_0, Omega1_0 = 0, Omega2_0 = 0, Tmax = 1000, dt = 0.001, ratio = 100):
    theta1, theta2, Omega1, Omega2 = simulate(theta1_0, theta2_0, Omega1_0, Omega2_0, Tmax, dt)
    return dropFrame(theta1, ratio), dropFrame(theta2, ratio), dropFrame(Omega1, ratio), dropFrame(Omega2, ratio)
```

```python
theta1, theta2, Omega1, Omega2 = simulateDropped(theta1_0 = np.pi/3, theta2_0 = np.pi/3, Tmax = 100, dt = 0.001, ratio = 100) 
ani = animate_double_pendulum((theta1, theta2, Omega1, Omega2), Tmax = 100)
```
